Trade Analysis – India and Global Perspective
India:
India's trade landscape is characterized by a diverse export base including petroleum products, pharmaceuticals, textiles, gems & jewelry, and IT services. Major trading partners include the USA, China, UAE, and the EU. Key import items are crude oil, gold, electronics, and machinery. In recent years, India has focused on reducing trade deficits, boosting exports through Production-Linked Incentive (PLI) schemes, and strengthening regional trade agreements.

Globally:
Global trade has been shaped by economic recovery post-COVID-19, supply chain realignments, geopolitical tensions, and technological shifts. Major players include the USA, China, Germany, and Japan. Trade in services, digital goods, and green technology is gaining importance. Organizations like the WTO and regional blocs play a key role in trade facilitation and dispute resolution. The push for sustainable and resilient supply chains is reshaping global trade dynamics.

**CAGR of India’s exports from 2015 to 2022.**


```python
# Filter only India
df_exports = df[['Reporter', 'Indicator', 'Year', 'Value']]

# Filter for India's exports
india_exports = df[(df['Reporter'] == 'India') & (df['Indicator'] == 'Exports')]



# Check if the filtered DataFrame is not empty before accessing
if not india_exports[india_exports['Year'] == 2015].empty and not india_exports[india_exports['Year'] == 2022].empty:
    start_value = india_exports[india_exports['Year'] == 2015]['Value'].values[0]
    end_value = india_exports[india_exports['Year'] == 2022]['Value'].values[0]
    years = india_exports[india_exports['Year'] == 2022]['Year'].values[0] - india_exports[india_exports['Year'] == 2015]['Year'].values[0]

    # Calculate CAGR
    n = 2022 - 2015
    cagr = (end_value / start_value) ** (1 / n) - 1
    print(f"CAGR: {cagr * 100:.2f}%")
else:
   print("Export data for 2015 or 2022 is missing.")
   print("❌ Missing years for India: [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]")



```

**CAGR of India's Export from 2015-2022 is 6.74%**

```python
# Filter data for:
# - Country
# - Year = 2022
# - Indicator = 'Import' (i.e., imports INTO India)

india_imports_2022 = df[
    (df['Reporter'] == 'India') &
    (df['Year'] == 2022) &
    (df['Indicator'] == 'Imports')]



# Group by Partner to sum export values (which is import for partner)
partner_imports = india_imports_2022.groupby('Partner')['Value'].sum().reset_index()



```

```python
print(partner_imports.empty)  # True means no data
print(partner_imports)
```

```python
print(partner_imports['Value'].isnull().sum())
```

```python
if not partner_imports.empty and partner_imports['Value'].notnull().any():
    max_partner = partner_imports.loc[partner_imports['Value'].idxmax()]
    print(f"Country importing the most from India in 2022: {max_partner['Partner']}")
    print(f"Import value: {max_partner['Value']}")
else:
    print("No import data available for India in 2022.")
```

```python
# Filter for India, export indicator, machinery HSN 84, years 2015-2022
india_machinery_exports = df[
    (df['Reporter'] == 'India') &
    (df['Indicator'] == 'Exports') &
    (df['ProductCode'] == 'MAMT') &  # HSN Code 84
    (df['Year'] >= 2015) &
    (df['Year'] <= 2022)
]

export_trend = india_machinery_exports.groupby('Year')['Value'].sum().reset_index()


# Plotting
plt.figure(figsize=(10,6))
plt.plot(export_trend['Year'], export_trend['Value'], marker='o', linestyle='-')
plt.title("India’s Export Trend for Machinery (HSN Code 84) from 2015–2022")
plt.xlabel("Year")
plt.ylabel("Export Value")
plt.grid(True)
plt.xticks(export_trend['Year'])
plt.tight_layout()
plt.show()
```

**📉 Export Trend Analysis for Machinery (Code: MAMT) – India (2015–2022)**

Recession Fears: Many economies entered recession or significant contraction in 2020, leading to lower demand for capital goods like machinery.
